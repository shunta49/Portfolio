```python
import numpy as np
import matplotlib.pyplot as plt

#情報から遷移行列Pを作成
P = np.array([[0.6, 0.3, 0.1], [0.3, 0.3, 0.4], [0.2, 0.5, 0.3]])
w = np.array([3/6, 2/6, 1/6])

Num = 20
w0 = np.array([0.2, 0.4, 0.4])
wTrack = np.zeros((3,Num))
wTrack[:,0] = w0
for k in range(1,Num):
    w = np.dot(w,P)
    wTrack[:,k] = w

#結果の表示
print(w)
plt.plot(wTrack[0,:],label='sunny',color='orange')
plt.plot(wTrack[1,:],label='cloudy',color='green')
plt.plot(wTrack[2,:],label='rainy',color='blue')
plt.xlabel('k')
plt.ylabel('w')
plt.grid()
plt.legend()
plt.show()
```
