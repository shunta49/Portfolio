```python
import numpy as np
import random
import matplotlib.pyplot as plt

N = 1000000
incircle = []
outcircle = []
for i in range(N):
   x = random.uniform(-1.0, 1.0)
   y = random.uniform(-1.0, 1.0)

   if x*x+y*y < 1.0:
     incircle.append([x,y])
   else:
     outcircle.append([x,y])
    
# 円周率の近似解
pi = 4.0 * len(incircle) / N

# 結果を表示 
print(pi)
incircle = np.array(incircle)
outcircle = np.array(outcircle)
plt.scatter(incircle[:,0], incircle[:,1], s=2, c='blue')
plt.scatter(outcircle[:,0], outcircle[:,1], s=2, c='red')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
